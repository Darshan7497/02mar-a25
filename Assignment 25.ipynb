# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a plotting library for the Python programming language that provides a range of tools for creating visualizations such as line plots, scatter plots, bar plots, histograms, and many more.



1.Support for various output formats such as PNG, PDF, SVG, etc.

2.Easy customization of plots with a wide range of styles, colors, and fonts.

3.Integration with NumPy, Pandas, and other Python libraries.

4.Interactive plotting capabilities with the use of widgets and GUI toolkits.




PLOTS IN MATPLOTLIB

Line Plot - It is used to display the relationship between two continuous variables.

Scatter Plot - It is used to display the relationship between two continuous variables, with each data point represented as a point on the plot.

Bar Plot - It is used to display the relationship between a categorical variable and a continuous variable.

Histogram - It is used to display the distribution of a continuous variable.

Pie Chart - It is used to display the proportion of different categories in a dataset.


# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

#Note: Also add title, xlabel, and ylabel to the plot.
```

```python
plt.scatter(x,y)
plt.xlabel("X AXIS")
plt.ylabel("Y AXIS")
plt.title("TITLE")
plt.show()
```

The subplot() function is used in Matplotlib library to create multiple subplots within a single figure. It allows us to display multiple plots side-by-side or stacked on top of each other, and is especially useful when we want to compare multiple visualizations or when we want to display different aspects of a dataset in separate plots.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x, y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x, y)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x, y)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x, y)
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


A bar plot is a type of data visualization that displays data using rectangular bars. The length of each bar corresponds to the value of a variable, and the bars are typically arranged along the x-axis with the y-axis representing the values. Bar plots are often used to compare the values of different categories or groups.

Bar plots are useful for several reasons. They can help to identify patterns and trends in data, as well as to highlight differences between groups or categories.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.barh(company,profit)
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


A box plot is a type of data visualization that displays the distribution of a dataset. It uses a rectangular box and whiskers to represent the interquartile range of the data and to identify potential outliers. Box plots are used to compare data across different groups or categories and to identify patterns and trends in data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2])
```
